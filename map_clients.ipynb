```python
import requests
import folium
import pandas as pd
from dotenv import load_dotenv
import os
import googlemaps
```

```python
load_dotenv()
API_KEY = os.getenv('GOOGLE_API_KEY')
gmaps = googlemaps.Client(key=API_KEY)
```

```python
hs_raw = pd.read_csv('data/hs_addresses.csv')
hs_customers = hs_raw[hs_raw['Status']=='Customer']
hs_locations = hs_customers.query("`Record Type` == 'Site' and `Street Address` == `Street Address`")
hs_locations.reset_index(inplace=True, drop=True)
```

```python
hs_locations.head()
```

```python
# Get coordinates for each school

for i in range(len(hs_locations)):
    try:
        hs_locations.loc[i, 'lat'] = gmaps.geocode(hs_locations['Street Address'][i])[0]['geometry']['location']['lat']
        hs_locations.loc[i, 'long'] = gmaps.geocode(hs_locations['Street Address'][i])[0]['geometry']['location']['lng']
    except IndexError:
        print(f"{hs_locations['Company name'][i]} at location {hs_locations['Street Address'][i]} not found")
```

```python
hs_locations.head()
```

```python
m = folium.Map(location=[39.8283, -98.5795], zoom_start=4)
```

```python
# Define a function to select color based on segment value
def get_color(segment):
    if segment == 'Allied Health':
        return 'blue'
    elif segment == 'Career + Trade':
        return 'orange'
    elif segment == 'Beauty + Wellness':
        return 'purple'
    # Add more conditions for different segments
    else:
        return 'green'  # Default color if segment doesn't match any condition

# Loop through each school and add a marker to the map
for i in range(len(hs_locations)):
    # Get the color for the marker based on the segment
    marker_color = get_color(hs_locations['Segment'][i])
    
    # Add marker to the map
    folium.Marker(
        location=[hs_locations['lat'][i], hs_locations['long'][i]], 
        popup=hs_locations['Company name'][i],
        icon=folium.Icon(color=marker_color)
    ).add_to(m)
```
